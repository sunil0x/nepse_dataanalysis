```python
#importing the pandas library
import pandas as pd
```

```python
df = pd.read_csv('../data-companywise/raw_data/MNBBL.csv')
df
```

```python
df.drop(columns = 'status', inplace = True)
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df=df[df['date'].dt.year >= 2022].copy()
df.to_csv('NABIL.csv', index= False)
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df=df[df['date'].dt.year >= 2022].copy()
df.to_csv('NABIL.csv', index= False)
```

```python
df
```

```python

#adding some fundamental signals
fundamental_data = {
    'year': [2022, 2023, 2024, 2025],
    'eps': [29.61, 23.36, 18.70, 16.80],  # NPR (Annualized from Q1/Q2 2082/83)
    'pe-ratio': [15.20, 16.50, 19.52, 21.55], 
    'roe': [16.82, 13.90, 13.50, 11.10],  # Statutory ROE %
    'dividend': [18.5, 14.2,10.2,18.2] # 2024 Actual: 18.2% (13.53% Bonus)
}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# #merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python

df = df.copy()
df.to_csv('../data-companywise/processed_data/MNBBL.csv', index = False)
```
