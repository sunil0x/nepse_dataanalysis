```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../data-companywise/raw_data/BPCL.csv')
df
```

```python
#dropping status column
df.drop(columns = 'status', inplace = True)
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df=df[df['date'].dt.year >= 2022].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
#adding some fundamental signals
fundamental_data = {
    'year': [2022, 2023, 2024, 2025],
    'eps': [8.05, 8.24, 8.13, 9.95],      # Highly stable income from PPA
    'pe-ratio': [38.20, 41.10, 52.40, 78.57], # P/E spiked in 2025 due to market speculation
    'roe': [4.50, 4.70, 4.30, 4.80],      # Very low ROE due to huge reserves
    'dividend': [12.50, 5.00, 5.00, 6.00]
}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# #merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../data-companywise/processed_data/BPCL.csv', index = False)
```
