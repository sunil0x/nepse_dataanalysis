```python
#importing the libraries
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../../data-companywise/raw_data/SCB.csv')
df
```

```python
df.drop(columns = 'status', inplace = True)
df
```

```python

# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df=df[df['date'].dt.year >= 2022].copy()
```

```python
#resetting index
df.reset_index(drop = True,  inplace= True)
df
```

```python
df['year'] = df['date'].dt.year
df
```

```python
#adding the fundamental data
fundamental_data = {
    'year':[2022,2023,2024,2025],
    'eps':[37.36, 43.19, 30.17, 27.35],
    'pe-ratio':[16.51, 14.80, 20.15, 23.03],
    'roe':[14.15, 17.55, 14.30, 12.96],
    'dividend':[16.51, 19.00, 25.50, 19.00],
    'net-margin': [39.19, 44.50, 47.36, 48.12],
    'pb-ratio': [1.85, 2.15, 2.40, 2.55],
    'debt-equity': [6.50, 6.20, 5.80, 5.50]
}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python

df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../../data-companywise/processed_data/SCB.csv', index = False)
```
