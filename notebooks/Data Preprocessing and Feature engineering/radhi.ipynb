```python
#importing pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../../data-companywise/raw_data/RADHI.csv')
df
```

```python
df.drop(columns = 'status', inplace = True)
df
```

```python

# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df
```

```python

df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df=df[df['date'].dt.year >= 2022].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
fundamental_data = {
    'year': [2022, 2023, 2024, 2025],
    'eps': [9.93, 7.22, 20.65, 60.58],  
    'pe-ratio': [45.2, 58.1, 37.4, 13.9], 
    'roe': [8.70, 6.30, 17.10, 26.57],   
    'dividend': [5.00, 5.00, 5.00, 5.00],
    'net-margin': [33.56, 32.92, 60.94, 63.84],
    'pb-ratio': [3.10, 3.40, 5.70, 5.50],
    'debt-equity': [1.12, 0.98, 0.65, 0.42]
}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# #merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../../data-companywise/processed_data/RADHI.csv', index = False)
```
