```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../../data-companywise/raw_data/SBL.csv')
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df.drop(columns = 'status', inplace = True)
df
```

```python
#year column created
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
#filtering the data from 2022 to 2025
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year > 2022].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python

fundamental_data = {
    'year': [2022, 2023, 2024, 2025], # FY 78/79, 79/80, 80/81, 81/82
    'eps': [21.80, 18.25, 12.45, 17.93],   # NPR (Annualized EPS)
    'pe-ratio': [14.20, 15.50, 24.60, 18.55], # Market price vs earnings
    'roe': [13.20, 11.50, 7.80, 13.96],    # Return on Equity (%)
    'dividend': [13.16, 4.21, 4.00, 10.53],# Total % (Bonus + Cash)
    'net-margin': [24.10, 21.20, 14.80, 22.40], # (%)
    'pb-ratio': [1.55, 1.45, 1.65, 1.87],      # Price-to-Book Ratio
    'debt-equity': [8.40, 8.65, 9.10, 8.95]}   # Bank leverage (Assets/Equity)

df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python

# merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../../data-companywise/processed_data/SBL.csv', index = False)
```
