```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../../data-companywise/raw_data/MEN.csv')
df
```

```python
df.drop(columns = 'status', inplace = True)
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df=df[df['date'].dt.year >= 2022].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
#adding some fundamental signals
fundamental_data = {
    'year': [2022, 2023, 2024, 2025], 
    'eps': [31.54, 38.21, 44.50, 54.24], 
    'pe-ratio': [18.72, 16.45, 13.27, 10.89], 
    'roe': [18.30, 21.10, 24.50, 26.80],  
    'dividend': [10.53, 15.79, 15.00, 21.05],
    'net-margin': [31.20, 32.50, 31.85, 33.40],
    'pb-ratio': [3.85, 4.10, 4.54, 4.35],
    'debt-equity': [1.35, 1.22, 1.15, 1.04]
}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# #merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python

df = df.copy()
df.to_csv('../../data-companywise/processed_data/MEN.csv', index = False)
```
