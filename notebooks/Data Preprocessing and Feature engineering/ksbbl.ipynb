```python
#importing the pandas library
import pandas as pd
```

```python

#reading the file
df = pd.read_csv('../../data-companywise/raw_data/KSBBL.csv')
df
```

```python
df.drop(columns = 'status', inplace = True)
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df=df[df['date'].dt.year >= 2022].copy()
```

```python
df
```

```python

#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
fundamental_data = {
    'year': [2022, 2023, 2024, 2025],
    'eps': [18.38, 11.01, 16.48, 20.43],  
    'pe-ratio': [18.50, 22.10, 24.12, 22.17], 
    'roe': [11.18, 7.63, 11.10, 12.57], 
    'dividend': [0.00, 4.63, 12.00, 15.79],
    'net-margin': [11.80, 10.45, 12.10, 12.70],
    'pb-ratio': [2.65, 2.85, 3.15, 3.10],
    'debt-equity': [9.40, 9.80, 9.55, 9.20]
}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# #merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../../data-companywise/processed_data/KSBBL.csv', index = False)
```
