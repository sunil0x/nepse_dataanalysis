```python

#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../../data-companywise/raw_data/NMB.csv')
df
```

```python
df.drop(columns = 'status', inplace = True)
df
```

```python

# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
#year column created
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year > 2022].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python

fundamental_data = {
    'year': [2022, 2023, 2024, 2025], # FY 78/79, 79/80, 80/81, 81/82
    'eps': [17.02, 12.12, 10.45, 11.20],   # Adjusted for capital base
    'pe-ratio': [15.10, 18.20, 21.45, 19.80], 
    'roe': [11.65, 8.12, 6.45, 7.50],    # Showing slight recovery in 2025
    'dividend': [8.25, 0.00, 0.00, 10.00], # Corrected: 10% declared in 2025
    'net-margin': [14.15, 9.05, 8.20, 9.50], 
    'pb-ratio': [1.38, 1.25, 1.20, 1.28],   
    'debt-equity': [8.10, 8.45, 8.70, 8.85]} 

df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../../data-companywise/processed_data/NMB.csv', index = False)
```
