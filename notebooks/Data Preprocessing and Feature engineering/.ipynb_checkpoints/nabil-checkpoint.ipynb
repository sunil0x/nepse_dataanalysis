```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('..//../data-companywise/raw_data/NABIL.csv')
df
```

```python
#removing the status column
df.drop(columns ='status',inplace= True)
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df=df[df['date'].dt.year >= 2022].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
#adding some fundamental signals
fundamental_data = {
    'year': [2022, 2023, 2024, 2025], 
    'eps': [37.42, 30.25, 25.84, 25.97],  
    'pe-ratio': [21.50, 19.80, 19.13, 19.13], 
    'roe': [14.80, 11.60, 10.50, 11.20],  
    'dividend': [30.00, 11.00, 10.00, 12.50], 
    'net-margin': [14.46, 15.25, 17.61, 22.02],
    'pb-ratio': [2.10, 3.20, 2.20, 2.35],
    'debt-equity': [9.10, 9.50, 9.80, 8.50]
}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# #merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python

df = df.copy()
df.to_csv('../../data-companywise/processed_data/NABIL.csv', index = False)
```
