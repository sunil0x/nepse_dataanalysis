```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../../data-companywise/raw_data/SHL.csv')
```

```python
df
```

```python
df.drop(columns = 'status', inplace = True)
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
#year column created
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year > 2022].copy()
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
df['year'] = df['date'].dt.year
df
```

```python
#adding some fundamental signals
fundamental_data = {
    'year':[2022,2023,2024,2025],
    'eps':[1.95,5.37,6.47,6.56],
    'pe-ratio':[98.4,82.5,78.2,81.7],
    'roe':[11.5,23.4,26.6,25.6],
    'dividend':[26.3,31.5,36.8,31.6]
}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df.tail(200)
```

```python
df
```

```python
df = df.copy()
df.to_csv('../../data-companywise/processed_data/SHL.csv', index = False)
```
