```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../../data-companywise/raw_data/SHEL.csv')
df
```

```python
#dropping the status column
df.drop(columns = 'status', inplace = True)
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df
```

```python
#year column created
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year > 2022].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python

fundamental_data = {
    'year': [2022, 2023, 2024, 2025],      # FY end periods (78/79 to 81/82)
    'eps': [3.12, 4.45, 6.70, 13.58],      # NPR (Growth post-commercial operation)
    'pe-ratio': [92.0, 75.4, 47.4, 24.8],  # Valuation softening as earnings rise
    'roe': [2.85, 3.90, 6.33, 11.78],      # Return on Equity (%)
    'dividend': [0.00, 0.00, 0.00, 5.00],  # 5% Bonus Share proposed for FY 2081/82
    'net-margin': [12.4, 18.2, 24.5, 32.1], # Operational efficiency improvement
    'pb-ratio': [3.20, 3.00, 2.90, 2.93],  
    'debt-equity': [2.10, 1.85, 0.85, 0.72] # Sharp drop due to 1:1 Right Share
}

df_fundamental = pd.DataFrame(fundamental_data)
print(df_fundamental)
```

```python
# #merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../../data-companywise/processed_data/SHEL.csv', index = False)
```
