```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../../data-companywise/raw_data/APi.csv')
df
```

```python

# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df
```

```python
#year column created
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python

#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year > 2022].copy()
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
df.drop(columns = 'status', inplace = True)
df
```

```python

fundamental_data = {
    'year': [2022, 2023, 2024, 2025], # FY 78/79 to 81/82
    'eps': [4.85, 3.63, 7.51, 14.03],    # NPR (Annualized EPS, adjusted for capital dilution)
    'pe-ratio': [38.50, 42.10, 35.80, 24.15], # Market price volatility vs earnings growth
    'roe': [4.52, 3.45, 6.88, 12.15],    # Return on Equity (%)
    'dividend': [7.50, 0.00, 5.00, 5.26], # Total % (Bonus + Cash)
    'net-margin': [22.45, 18.12, 36.50, 44.50], # (%)
    'pb-ratio': [1.85, 1.95, 2.45, 2.75],     # Price-to-Book Ratio
    'debt-equity': [1.15, 1.58, 0.90, 0.85]} # Gearing ratio (Total Debt/Equity)

df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../../data-companywise/processed_data/API.csv', index = False)
```
