```python
#importing the pandas library
import pandas as pd
```

```python

#reading the file
df = pd.read_csv('../../data-companywise/raw_data/GBBL.csv')
df
```

```python
df.drop(columns = 'status', inplace = True)
df
```

```python

# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
#year column created
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year > 2022].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python

fundamental_data = {
    'year': [2022, 2023, 2024, 2025], # FY 78/79 to 81/82
    'eps': [21.56, 17.02, 7.64, 21.10],   # NPR (Audited & 2025 Annualized)
    'pe-ratio': [17.50, 22.30, 48.55, 19.93], # Based on price (~420 as of Mar 2026)
    'roe': [14.10, 10.85, 4.63, 13.22],   # Return on Equity (%)
    'dividend': [14.50, 10.00, 5.00, 10.53], # Total % (Bonus + Cash)
    'net-margin': [24.12, 19.45, 10.58, 26.35], # (%)
    'pb-ratio': [2.10, 2.05, 2.38, 2.63],
    'debt-equity': [9.45, 9.80, 10.12, 10.30]} # Leverage (Total Assets/Equity proxy)

df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../../data-companywise/processed_data/GBBL.csv', index = False)
```
