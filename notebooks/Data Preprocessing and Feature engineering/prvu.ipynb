```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../../data-companywise/raw_data/PRVU.csv')
df
```

```python
df.drop(columns = 'status', inplace = True)
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
#year column created
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
#filtering the data from 2022 to latest
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year > 2022].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python

fundamental_data = {
    'year': [2022, 2023, 2024, 2025], # FY 78/79, 79/80, 80/81, 81/82
    'eps': [17.85, 4.41, 2.17, 11.35],    # NPR (Reflects post-merger dilution and recovery)
    'pe-ratio': [12.50, 28.50, 45.10, 18.20], # Market adjusted
    'roe': [11.20, 3.10, 1.45, 7.85],      # Reflects integration-driven volatility
    'dividend': [8.00, 0.00, 0.00, 0.00],  # No dividends declared during integration/retention
    'net-margin': [18.20, 8.40, 5.20, 10.12], # (%)
    'pb-ratio': [0.95, 0.85, 0.75, 0.82],     # Trading below book value
    'debt-equity': [9.50, 10.20, 10.85, 10.15] # Leverage
}

df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../../data-companywise/processed_data/PRVU.csv', index = False)
```
