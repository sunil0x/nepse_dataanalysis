```python
#importing the pandas library
import pandas as pd
```

```python

#reading the file
df = pd.read_csv('../../data-companywise/raw_data/SAHAS.csv')
df
```

```python
df.drop(columns = 'status', inplace = True)
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df
```

```python

#year column created
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df=df[df['date'].dt.year >= 2022].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
fundamental_data = {
    'year': [2022, 2023, 2024, 2025], # Corresponding to FY 78/79, 79/80, 80/81, 81/82
    'eps': [0.18, 2.94, 9.10, 23.88],
    'pe-ratio': [3245.0, 185.20, 68.50, 28.50],
    'roe': [0.12, 1.84, 6.94, 13.63],
    'dividend': [0.00, 0.00, 8.42, 22.11],
    'net-margin': [16.99, 26.36, 34.03, 40.04],
    'pb-ratio': [2.80, 3.20, 3.70, 3.40],
    'debt-equity': [2.10, 1.85, 1.60, 1.45]
}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../../data-companywise/processed_data/SAHAS.csv', index = False)
```
