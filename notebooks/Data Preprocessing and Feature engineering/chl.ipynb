```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../../data-companywise/raw_data/CHL.csv')
df
```

```python
#dropping the status column
df.drop(columns = 'status', inplace = True)
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
df
```

```python
#year column created
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year > 2022].copy()
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python

fundamental_data = {
    'year': [2022, 2023, 2024, 2025],      # FY end periods
    'eps': [-2.35, -4.50, -6.42, -5.80],  # NPR (Reported, negative due to losses)
    'pe-ratio': [0.00, 0.00, 0.00, 0.00],  # N/A due to negative earnings
    'roe': [-3.8, -7.2, -9.5, -8.2],       # Return on Equity (%)
    'dividend': [0.00, 0.00, 0.00, 0.00],  # No dividends due to accumulated losses
    'net-margin': [-42.0, -44.0, -54.0, -36.0], # Operating losses
    'pb-ratio': [2.8, 3.5, 4.0, 3.7],      # Market valuation relative to book value
    'debt-equity': [2.1, 2.9, 3.5, 5.1]    # High leverage prior to 2025 right share
}

df_fundamental = pd.DataFrame(fundamental_data)
print(df_fundamental)
```

```python
# #merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../../data-companywise/processed_data/CHL.csv', index = False)
```
