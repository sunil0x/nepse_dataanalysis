```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../data-companywise/CIT.csv')
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df
```

```python
#year column created
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year.between(2022, 2025)].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
df.drop(columns = 'status', inplace = True)
df
```

```python
fundamental_data = {
    'year': [2022, 2023, 2024, 2025], # Corresponding to FY 78/79, 79/80, 80/81, 81/82
    'eps': [20.30, 17.08, 16.20, 19.40],
    'pe-ratio': [125.40, 122.27, 131.30, 95.43],
    'roe': [8.87, 9.40, 12.61, 12.67],
    'dividend': [26.43, 14, 13.74, 13.00]}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../data-companywise/CIT.csv', index = False)
```
