```python
#importing the pandas library
import pandas as pd
```

```python
csv_path = "S:/data science project/nepse_dataanalysis/share_data/SHL.csv"
SHL_df = pd.read_csv(csv_path)
```

```python
SHL_df
```

```python
# Just call the function; do not use the "=" sign
SHL_df.drop(columns=['status'], inplace=True)
SHL_df
```

```python
#renaming the published_date to date
# Rename specific columns to the "Perfect Schema"
SHL_df = SHL_df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
#filtering the data from 2022
SHL_df['date'] = pd.to_datetime(SHL_df['date'])
SHL_df=SHL_df[SHL_df['date'].dt.year >= 2022].copy()
SHL_df.to_csv('SHL.csv', index= False)

```

```python
SHL_df
```

```python
# Rename specific columns to the "Perfect Schema"
SHL_df.reset_index(drop= True, inplace = True)
SHL_df
```

```python
#adding the column
df = SHL_df.copy()
df
```

```python
df['year'] = df['date'].dt.year
df
```

```python
#adding some fundamental signals
fundamental_data = {
    'year':[2022,2023,2024,2025],
    'eps':[1.95,5.37,6.47,6.56],
    'pe-ratio':[98.4,82.5,78.2,81.7],
    'roe':[11.5,23.4,26.6,25.6],
    'dividend':[26.3,31.5,36.8,31.6]
}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```
