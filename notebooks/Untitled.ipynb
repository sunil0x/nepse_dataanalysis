```python
import pandas as pd
import glob
import os

csv_files = glob.glob("share_data/*.csv")
print(csv_files)
```

```python
for file in csv_files:
    print("Processing:", file)
    df = pd.read_csv(file)
    df.drop(columns=["status"], inplace=True, errors="ignore")
    df.to_csv(file, index=False)
print('done')
```

```python
for file in csv_files:
    df = pd.read_csv(file)
    df.rename(columns={'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'})
    df.to_csv(file, index=False)
    
```
