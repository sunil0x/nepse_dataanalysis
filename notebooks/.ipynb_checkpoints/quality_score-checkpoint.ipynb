# Quality score- Performance of the company analysis

1. Data collection and data preprocessing

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#taking the input to ask the user to choose the company
name = input("Enter the name of the company").upper()
name

#taking the dataset 
df = pd.read_csv(f"../data-companywise/{name}.csv")
df
```

```python
# df = df.drop(df.columns[0], axis=1)
```

```python
df = df[(df['year'] >= 2022 ) & (df['year'] < 2026)].copy()
```

```python
# df.to_csv(f"../data-companywise/{name}.csv")
```

```python
df
```

```python
#checking the null value
df.isna().sum()
```

```python
df.info()
```

2. Data Preprocessing

```python
#setting date as index 
df.set_index('date',inplace = True)
df
```

```python
#sorting the date
df.sort_index(inplace = True)
df
```

3. Feature Engineering

```python
# Calculating the daily percentage change(returns) and volatility
df['price_change_pct'] = df['close'].pct_change().fillna(0)
df['volatility'] = df['price_change_pct'].rolling(window=2).std().fillna(0)
```

```python
# 4. Clean up
# We drop NaNs because: 10 from volatility contains null, 
df.dropna(inplace=True)
```

```python
df
```

4. Quality score

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
#Quality Score Calculation (The Formula)
scaler = MinMaxScaler()
# Higher is better
pos = ['eps', 'roe', 'dividend', 'close', 'volume', 'price_change_pct']
df[[f'{c}_score' for c in pos]] = scaler.fit_transform(df[pos])

#lower is better
df['pe_score'] = 1 - scaler.fit_transform(df[['pe-ratio']])
df['vol_score'] = 1 - scaler.fit_transform(df[['volatility']])
```

```python
# Final Score
score_cols = [f'{c}_score' for c in pos] + ['pe_score', 'vol_score']
df['quality_score'] = df[score_cols].mean(axis=1)
```

```python
df
```

evolution graph of quality score

```python
#quality score graph
import matplotlib.pyplot as plt
#quality score graph
plt.figure(figsize = (10,6))
plt.plot(df['year'], df['quality_score'],color='#6a0dad', linewidth=2, label='Quality Score')
plt.title(' Fundamental Quality Evolution', fontsize=16, fontweight='bold')
plt.xlabel('Date (2022 - 2026)', fontsize=12)
plt.ylabel('Quality Score (Normalized 0 to 1)', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle=':', alpha=0.5)
plt.show()
```

# modeling the random forest model

```python
# 4. Random Forest (The Algorithm)
# We create a 'target' so the machine can learn: 1 = Improvement, 0 = Decline
df['target'] = (df['quality_score'].diff() > 0).astype(int)
```

```python
df
```

Feature selection

```python
#splitting the data into the train and test
from sklearn.model_selection import train_test_split
features = ['close', 'volume', 'eps', 'pe-ratio', 'roe', 'dividend', 'price_change_pct', 'volatility']
X = df[features]
y = df['quality_score']

#here 25 % for the test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

initialize and apply random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# 3. Use RandomForestRegressor to output the score
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# 4. Generate Predictions
# The output here will be the predicted Quality Score
predicted_scores = rf_model.predict(X_test)
```

```python
#Feature Importance
importances = pd.Series(rf_model.feature_importances_, index=features).sort_values(ascending=False)
print("\nWhich metric defines the Quality Score the most?")
print(importances)
```

```python
#calculating the errors
mean_square_error = mean_squared_error(y_test, predicted_scores)
r2_err = r2_score(y_test, predicted_scores)
print(f"Model R^2 Score:{r2_err:.4f} ")
print(f"Mean squared Error: {mean_square_error:.4f}")
```

visualizing the importance 

```python
# 5. Extract Feature Importances
importances = rf_model.feature_importances_
feature_df = pd.DataFrame({'Feature': features, 'Importance': importances})

# Sorting for a cleaner bar chart (Ascending for horizontal plot)
feature_df = feature_df.sort_values(by='Importance', ascending=True)

# Plotting
plt.barh(feature_df['Feature'], feature_df['Importance'], color='teal')
plt.title('Which Metrics Drive the Quality Score?', fontsize=14, fontweight='bold')
plt.xlabel('Importance Weight')
plt.tight_layout() # Ensures labels like 'price_change_pct' are not cut off
plt.savefig('feature_importance.png')
```

```python
#The close vs. Quality Score (The "Penalty" View)
plt.figure(figsize=(10, 6))
plt.scatter(df['close'], df['quality_score'], alpha=0.5, c=df['quality_score'], cmap='RdYlGn')
plt.axvline(x=50, color='red', linestyle='--', label='High P/E Threshold') # Example threshold
plt.xlabel('closing price')
plt.ylabel('Quality Score')
plt.title('Quality Score vs. closing price')
plt.xscale('log') # Useful if you have values like 3245
plt.colorbar(label='Quality Score Intensity')
plt.legend()
plt.show()
```

```python
import seaborn as sns

# Selecting the requested metrics for the correlation matrix
correlation_cols = ['quality_score', 'eps', 'roe', 'dividend', 'pe-ratio', 'volatility']
correlation_matrix = df[correlation_cols].corr()

# Plotting the Heatmap
plt.title('Correlation between Quality and Fundamentals', fontsize=14, fontweight='bold')
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
```

```python
#Predicted vs. Actual Quality (Model Performance)
plt.figure(figsize=(8, 8))
plt.scatter(y_test, predicted_scores, alpha=0.6, color='purple')
plt.plot([0, 1], [0, 1], 'r--') # Perfect prediction line
plt.xlabel('Actual Quality Score')
plt.ylabel('Predicted Quality Score')
plt.title('Random Forest Accuracy: Actual vs Predicted')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

# 1. Define Features (Including SMA as requested)
# Note: With yearly data, SMA-5 is calculated using a shorter window to fit the sample size
df['sma-5'] = df['close'].rolling(window=2).mean().fillna(df['close'])
features = ['close', 'eps', 'roe', 'dividend', 'pe-ratio', 'volatility']

# 2. Create the 'Forecast Target' (Shift -1)
# This aligns current year's features with next year's result
df['target_quality'] = df['quality_score'].shift(-1)

# 3. Prepare for Training (Remove the last row which has no 'future' score)
df_model = df.dropna(subset=['target_quality']).copy()
X = df_model[features]
y = df_model['target_quality']

# 4. Train the Random Forest Regressor
rf_forecast = RandomForestRegressor(n_estimators=100, random_state=42)
rf_forecast.fit(X, y)

# 5. Generate Predictions for Visualization
df_model['predicted_quality'] = rf_forecast.predict(X)
```

```python
# 4. Final Visualization
plt.plot(df_model['year'], df_model['quality_score'], label='Current Fundamental Quality', color='gray', alpha=0.4, linewidth=2, marker='o')
plt.plot(df_model['year'], df_model['predicted_quality'], label='Forecasted Quality (Dynamic)', color='purple', linewidth=1.5, marker='x')

plt.title('Dynamic Quality Forecast: How "Close" Price Influences Quality Prediction', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Quality Score')
plt.legend()
plt.tight_layout()
```

```python
latest_day_features = df[features].tail(1)

# 2. CALCULATE the upcoming score
# We use the trained rf_dynamic regressor to project the next score
upcoming_score = rf_forecast.predict(latest_day_features)[0]

# 3. Get the current quality score to compare against
# This is the actual score calculated from the most recent 2025 data
current_score = df['quality_score'].iloc[-1]
```

```python
# 4. Generate the Investment Signal (for your final cell)
upcoming_rounded = round(upcoming_score, 4)
current_rounded = round(current_score, 4)

if upcoming_rounded > current_rounded:
    signal = "BULLISH (Quality Improving)"
elif upcoming_rounded < current_rounded:
    signal = "BEARISH (Quality Declining)"
else:
    signal = "NEUTRAL (Quality Stable)"

print(f"Current Quality (2025): {current_rounded}")
print(f"Predicted Upcoming Quality: {upcoming_rounded}")
print(f"Signal: {signal}")
```
