```python

#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../data-companywise/HBL.csv')
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python

df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df=df[df['date'].dt.year >= 2022].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
fundamental_data = {
    'year': [2022, 2023, 2024, 2025], # FY 78/79 to 81/82
    'eps': [21.85, 16.34, 10.95, 11.45],  # NPR (Actual Audited & Q2 Annualized)
    'pe-ratio': [14.55, 11.87, 18.15, 16.76], # Precise P/E based on Jan 2026 price (~191)
    'roe': [12.10, 9.40, 6.27, 6.57],     # Return on Equity (%)
    'dividend': [19.11, 0.00, 0.00, 0.00]  # TOTAL % (The "Zero Dividend" period)
}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# #merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../data-companywise/HBL.csv', index = False)
```
