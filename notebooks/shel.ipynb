```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../data-companywise/raw_data/SHEL.csv')
df
```

```python
#dropping the status column
df.drop(columns = 'status', inplace = True)
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df
```

```python
#year column created
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year.between(2022, 2025)].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
fundamental_data_shel_actual = {
    'year': [2022, 2023, 2024, 2025], 
    'eps': [0.13, -2.27, -0.31, 5.17],
    'pe-ratio': [1250.0, -112.50, -817.0, 61.20],
    'roe': [0.15, -2.42, -0.33, 4.89],
    'dividend': [0.00, 0.00, 0.00, 5.00]}

df_shel = pd.DataFrame(fundamental_data_shel_actual)
df_shel
```

```python

# merging
df = pd.merge(df, df_shel, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../data-companywise/processed_data/SHEL.csv', index = False)
```
