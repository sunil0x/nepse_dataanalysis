```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../data-companywise/ADBL.csv')
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python

fundamental_data = {
    'year': [2022, 2023, 2024, 2025], # FY 78/79, 79/80, 80/81, 81/82
    'eps': [21.85, 10.32, 18.25, 23.44],  # Adjusted for Preference Dividend
    'pe-ratio': [15.24, 34.60, 14.58, 12.63], # Based on Fiscal Year-end prices
    'roe': [12.10, 5.64, 9.88, 11.95],     # Return on Equity (%)
    'dividend': [13.00, 11.00, 10.53, 13.00]} # Total % (Bonus + Cash)

df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# #merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../data-companywise/ADBL.csv', index = False)
```
