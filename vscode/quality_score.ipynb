2. Data Preprocessing

```python
#setting date as index 
df.set_index('date',inplace = True)
df
```

```python
#sorting the date
df.sort_index(inplace = True)
df
```

3. Feature Engineering

```python
#technical data indicators( simple moving average)
df['sma-5'] = df['close'].rolling(window=5).mean()
df['sma-20'] = df['close'].rolling(window=20).mean()
```

```python
# Calculating the daily percentage change(returns)
df['price_change_pct'] = df['close'].pct_change() * 100
```

```python
# 2. Now calculate Daily Volatility (Standard Deviation over a 10-day window)
df['volatility'] = df['price_change_pct'].rolling(window=10).std()

# 3. Create the Target (Tomorrow's price)
df['target'] = df['close'].shift(-1)

# 4. Clean up
# We drop NaNs because: 10 from volatility contains null, 
df.dropna(inplace=True)
```

```python
df
```

4. Quality score

```python
#calculating the payout ratio
df['payout-ratio'] = df['dividend']/df['eps']

#grouping by the year
yearly_data = df.groupby('year')[['eps','roe','dividend']].mean()
# print(yearly_data)

#normalizing the value
#normalizing eps
eps_min, eps_max = df['eps'].min(), df['eps'].max()
df['eps_norm'] = (df['eps'] - eps_min) / (eps_max - eps_min)

#Normalizing ROE
roe_min, roe_max = df['roe'].min(), df['roe'].max()
df['roe_norm'] = (df['roe'] - roe_min) / (roe_max - roe_min)

#normalizing dividend
dividend_min, dividend_max = df['dividend'].min(), df['dividend'].max()
df['dividend_norm'] = (df['dividend'] - dividend_min) / (dividend_max - dividend_min)
```

```python
#final quality score
df['quality_score'] = (df['eps_norm'] * 0.4) + (df['roe_norm'] * 0.4) + (df['dividend_norm'] * 0.2)
```

```python
yearly_quality_score = df.groupby('year')[['quality_score']].mean()
yearly_quality_score
```

evolution graph of quality score

```python
#quality score graph
plt.figure(figsize = (10,6))
plt.plot(df['year'], df['quality_score'],color='#6a0dad', linewidth=2, label='Quality Score')
plt.title(' Fundamental Quality Evolution', fontsize=16, fontweight='bold')
plt.xlabel('Date (2022 - 2026)', fontsize=12)
plt.ylabel('Quality Score (Normalized 0 to 1)', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle=':', alpha=0.5)
plt.show()
```

Feature selection

```python
#splitting the data into the train and test
from sklearn.model_selection import train_test_split
features = ['eps', 'pe-ratio', 'roe', 'dividend', 'volume']
X = df[features]
y = df['quality_score']

#here 25 % for the test data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25,random_state = 42)
```

initialize and apply random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

#initialialize the model
rf_model = RandomForestRegressor(n_estimators = 100, random_state = 42)

#training the model
rf_model.fit(X_train, y_train)

#making the predictions
y_pred = rf_model.predict(X_test)
```

```python
#calculating the errors
mean_square_error = mean_squared_error(y_test, y_pred)
r2_err = r2_score(y_test, y_pred)
print(f"Model R^2 Score:{r2_err:.4f} ")
print(f"Mean squared Error: {mean_square_error:.4f}")
```

visualizing the importance 

```python
# 5. Extract Feature Importances
importances = rf_model.feature_importances_
feature_df = pd.DataFrame({'Feature': features, 'Importance': importances})
feature_df = feature_df.sort_values(by='Importance', ascending=True)

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(feature_df['Feature'], feature_df['Importance'], color='teal')
plt.title('Which Metrics Drive the Quality Score?', fontsize=14, fontweight='bold')
plt.xlabel('Importance Weight')
plt.show()
```

```python
#predicting the upcoming quality score
# 1. We create a new column: 'target_quality'
# 1. Add 'close' to your features
features = ['close', 'eps', 'roe', 'dividend', 'pe-ratio', 'volatility', 'sma-5']

# 2. Create the 'Forecast Target' (Shift the quality score by -1)
# This means: Target for Row 1 is the Quality Score of Row 2
df['target_quality'] = df['quality_score'].shift(-1)

# 3. Remove the last row (it has no 'tomorrow' to predict)
df_model = df.dropna(subset=['target_quality']).copy()

X = df_model[features]
y = df_model['target_quality']
```

```python
#training model
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# Split data (sequential split for time series)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Train the model
rf_dynamic = RandomForestRegressor(n_estimators=100, random_state=42)
rf_dynamic.fit(X_train, y_train)

# Generate predictions for the whole dataset to see the movement
df_model['predicted_quality'] = rf_dynamic.predict(X)
```

```python
#visualizing forecast
plt.figure(figsize=(14, 7))

# Actual Quality (the flat steps based on yearly reports)
plt.plot(df_model.index, df_model['quality_score'], label='Current Fundamental Quality', color='gray', alpha=0.4, linewidth=2)

# Predicted Quality (the dynamic movement including 'close' price sentiment)
plt.plot(df_model.index, df_model['predicted_quality'], label='Forecasted Quality (Dynamic)', color='purple', linewidth=1.5)

plt.title('Dynamic Quality Forecast: How "Close" Price Influences Quality Prediction', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Quality Score')
plt.legend()
plt.show()
```

```python
# 1. Prepare the data for the final day (Today)
# We use the 'features' list defined earlier: ['close', 'eps', 'roe', 'dividend', 'pe-ratio', 'volatility', 'sma-5']
latest_day_features = df[features].tail(1)

# 2. CALCULATE the upcoming score (This defines the missing variable)
# We use the 'rf_dynamic' model you trained in the previous cell
upcoming_score = rf_dynamic.predict(latest_day_features)[0]

# 3. Get the current quality score to compare against
current_score = df['quality_score'].iloc[-1]
```

```python
# 4. Now generate the Investment Signal
# Round to 4 decimal places to ignore tiny precision differences
upcoming_rounded = round(upcoming_score, 4)
current_rounded = round(current_score, 4)

if upcoming_rounded > current_rounded:
    signal = "BULLISH (Quality Improving)"
elif upcoming_rounded < current_rounded:
    signal = "BEARISH (Quality Declining)"
else:
    signal = "NEUTRAL (Quality Stable)"

print(f"Upcoming: {upcoming_rounded} | Current: {current_rounded}")
print(f"Signal: {signal}")
```
