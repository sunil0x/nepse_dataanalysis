```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../data-companywise/APi.csv')
df
```

```python

# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df
```

```python
#year column created
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python

#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year.between(2022, 2025)].copy()
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
# df.drop(columns = 'status', inplace = True)
# df
```

```python
fundamental_data = {
    'year': [2022, 2023, 2024, 2025],
    'eps': [3.67, 1.06, 8.79, 17.30],
    'pe-ratio': [66.62, 240.50, 33.90, 17.16],
    'roe': [4.02, 1.06, 7.54, 14.10],
    'dividend': [ 0.00,7.89, 5.26, 5.26]}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python
df = df.copy()
df.to_csv('../data-companywise/API.csv', index = False)
```
