```python
#importing the pandas library
import pandas as pd
```

```python
#reading the file
df = pd.read_csv('../data-companywise/MEN.csv')
df
```

```python
# Rename specific columns to the "Perfect Schema"
df = df.rename(columns={
    'published_date': 'date',
    'traded_quantity': 'volume',
    'traded_amount': 'turnover',
    'per_change': 'price_change_pct'
})
```

```python
df['date'] = pd.to_datetime(df['date'])   # convert first
df['year'] = df['date'].dt.year
```

```python
df
```

```python
#filtering the data from 2022
df['date'] = pd.to_datetime(df['date'])
df=df[df['date'].dt.year >= 2022].copy()
```

```python
df
```

```python
#resetting index
df.reset_index(drop= True, inplace = True)
df
```

```python
#adding some fundamental signals
fundamental_data = {
    'year': [2022, 2023, 2024, 2025], # Nepali FY 78/79 to 81/82
    'eps': [31.54, 38.21, 44.50, 54.24],  # NPR (Actual reported figures)
    'pe-ratio': [18.72, 16.45, 13.27, 10.89], # Note the falling P/E despite price growth
    'roe': [18.30, 21.10, 24.50, 26.80],  # ROE is high due to low maintenance costs
    'dividend': [10.53, 15.79, 15.00, 21.05]
}
df_fundamental = pd.DataFrame(fundamental_data)
df_fundamental
```

```python
# #merging
df = pd.merge(df, df_fundamental, on='year',how= 'left')
df
```

```python

df = df.copy()
df.to_csv('../data-companywise/MEN.csv', index = False)
```
